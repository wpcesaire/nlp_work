# News Headlines Sentiment

Use the news api to pull the latest news articles for bitcoin and ethereum and create a DataFrame of sentiment scores for each coin. 

Use descriptive statistics to answer the following questions:
1. Which coin had the highest mean positive score?
2. Which coin had the highest negative score?
3. Which coin had the highest positive score?

```python
# Read your api key environment variable
api_key = os.getenv("api_key")
```

```python
print(type(api_key))
```

```python
# Create a newsapi client
newsapi = NewsApiClient(api_key=api_key)
```

```python
print(newsapi)
```

```python
# Fetch the Bitcoin news articles
bitcoin_headlines = newsapi.get_everything(q='Bitcoin', language='en')

# Sum of articles
print(f"Total articles regarding Bitcoin \"BTC\": {bitcoin_headlines['totalResults']}")
```

```python
# Display sample, 6th article
print(bitcoin_headlines['articles'][5])
```

```python
# Fetch the Ethereum news articles
ethereum_headlines = newsapi.get_everything(q='ethereum AND +ethereum AND ETH', language='en')
```

```python
# Total Ethereum Articles fetched
print(f"Total articles regarding Ethereum \"ETH\": {ethereum_headlines['totalResults']}")
```

```python
# Sample 3rd article
print(ethereum_headlines['articles'][2])
```

```python
# Create the Bitcoin sentiment scores DataFrame
bitcoin_sentiments = []
for article in bitcoin_headlines["articles"]:
    try:
        text = article["title"]
        date = article["publishedAt"][:10]
        sentiment = analyzer.polarity_scores(text)
        compound = sentiment["compound"]
        pos = sentiment["pos"]
        neu = sentiment["neu"]
        neg = sentiment["neg"]
        
        bitcoin_sentiments.append({"text": text, "date": date, "compound": compound, "positive": pos, "negative": neg, "neutral": neu})
    except AttributeError:
        pass
    
# DataFrame
bitcoin_df = pd.DataFrame(data=bitcoin_sentiments)
    
# Reorder
cols = ["date", "text", "compound", "positive", "negative", "neutral"]
bitcoin_df = bitcoin_df[cols]
bitcoin_df
```

```python
# Create the ethereum sentiment scores DataFrame
ethereum_sentiments = []
for article in ethereum_headlines["articles"]:
    try:
        text = article["content"]
        date = article["publishedAt"][:10]
        sentiment = analyzer.polarity_scores(text)
        compound = sentiment["compound"]
        pos = sentiment["pos"]
        neu = sentiment["neu"]
        neg = sentiment["neg"]
        
        ethereum_sentiments.append({
            "text": text, "date": date, "compound": compound, "positive": pos, "negative": neg, "neutral": neu})
    except AttributeError:
        pass
# Create DataFrame
ethereum_df = pd.DataFrame(ethereum_sentiments)
# Reorder DataFrame columns
cols = ["date", "text", "compound", "positive", "negative", "neutral"]
ethereum_df = ethereum_df[cols]
ethereum_df.head(10)
```

```python
# Describe the Bitcoin Sentiment
bitcoin_df.describe()
```

```python
# Describe the Ethereum Sentiment
ethereum_df.describe()
```

### Questions:

Q: Which coin had the highest mean positive score?

A: Ethereum

Q: Which coin had the highest compound score?

A: Ethereum

Q. Which coin had the highest positive score?

A: Bitcoin

```python
# Complete the tokenizer function
def bitcoin_tokenizer(text):
    """Tokenizes text."""
    
    # Create a list of the words
    sw = set(stopwords.words('english'))
    sw_addon = {'share','introducing', 'announced', 'user', 'state', 'online', 'boomers', 'train', 'avocado'}
    
    # Substitute everything that is NOT a letter with empty string
    regex = re.compile("[^a-zA-Z ]")
    bitcoin_re_clean = regex.sub('', bitcoin_article)
        
    # Convert the words to lowercase
    bitcoin_re_words = word_tokenize(bitcoin_re_clean)
    bitcoin_re_result = [word.lower() for word in bitcoin_re_words if word.lower() not in sw.union(sw_addon)]
  
      
    # Lemmatize Words into root words
    bitcoin_words = word_tokenize(bitcoin_re_clean)
       
    bitcoin_output = [word.lower() for word in bitcoin_words if word.lower() not in sw.union(sw_addon)]
        
    
    return bitcoin_output
```

# NGrams and Frequency Analysis

In this section you will look at the ngrams and word frequency for each coin. 

1. Use NLTK to produce the n-grams for N = 2. 
2. List the top 10 words for each coin. 

```python
from collections import Counter
from nltk import ngrams
```

```python
# Generate the Bitcoin N-grams where N=2
bitcoin_bigram_counts = [Counter(ngrams(tokens, n=2)) for tokens in bitcoin_df.tokens]
dict(bitcoin_bigram_counts[0].most_common(10))
```

```python
# Generate the Ethereum N-grams where N=2
ethereum_bigram_counts = [Counter(ngrams(tokens, n=2)) for tokens in ethereum_df.tokens]
dict(ethereum_bigram_counts[0].most_common(10))
```

```python
# Use the token_count function to generate the top 10 words from each coin
def bitcoin_token_count(bitcoin_tokens, N=10):
    """Returns the top N tokens from the frequency count"""
    return Counter(bitcoin_tokens).most_common(N)
```

```python
def ethereum_token_count(ethereum_tokens, N=10):
    """Returns the top N tokens from the frequency count"""
    return Counter(ethereum_tokens).most_common(N)
```

```python
# Get the top 10 words for Bitcoin
bitcoin_big_string = []
for tokens in bitcoin_df.tokens:
    bitcoin_big_string += tokens
bitcoin_token_count = bitcoin_token_count(bitcoin_big_string)
bitcoin_token_count
```

```python
# Get the top 10 words for Ethereum
ethereum_big_string = []
for tokens in ethereum_df.tokens:
    ethereum_big_string += tokens
ethereum_token_count = ethereum_token_count(ethereum_big_string)
ethereum_token_count
```

# Named Entity Recognition

In this section, you will build a named entity recognition model for both coins and visualize the tags using SpaCy.

```python
import spacy
from spacy import displacy
```

```python
# Optional - download a language model for SpaCy
# !python -m spacy download en_core_web_sm
```

```python
# Load the spaCy model
nlp = spacy.load('en_core_web_sm')
```

## Bitcoin NER

```python
# Concatenate all of the bitcoin text together
all_bitcoin = []

for text in bitcoin_df['text']:
    all_bitcoin.append(text)
all_bitcoin = "".join(all_bitcoin)
all_bitcoin
```

```python
# Run the NER processor on all of the text
bitcoin_ner = nlp(all_bitcoin)

# Add a title to the document
bitcoin_ner.user_data["title"] = "Bitcoin NER"
```

```python
# Render the Bitcoin NER visualization
displacy.render(bitcoin_ner, style='ent')
```

```python
# List all Entities
[(ent.text, ent.label_) for ent in bitcoin_ner.ents]
```
